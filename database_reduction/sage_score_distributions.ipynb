```python
# Calculate FDR for F07 and S11

#F07
# Sort dataframe by score in descending order
psm_df_F_sorted = psm_df_F.sort_values('score', ascending=False)

# Calculate cumulative number of decoy and target hits
psm_df_F_sorted['cumulative_decoy'] = psm_df_F_sorted['is_decoy'].cumsum()
psm_df_F_sorted['cumulative_total'] = psm_df_F_sorted['score'].expanding().count()

# Calculate FDR
psm_df_F_sorted['FDR'] = psm_df_F_sorted['cumulative_decoy'] / psm_df_F_sorted['cumulative_total']

# Find score at which FDR first falls below or equals to 1%
fdr_1_percent_score_F = psm_df_F_sorted[psm_df_F_sorted['FDR'] <= 0.01]['score'].min()
```
