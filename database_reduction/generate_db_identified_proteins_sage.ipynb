```python
# Extract the relevant column
protein_list_F = psm_df_F['protein_list'].tolist()
```

```python
protein_list_S = psm_df_S['protein_list'].tolist()
```

```python
# Delete all duplicates from the protein list
unique_protein_list_F = [list(item) for item in dict.fromkeys(tuple(sublist) for sublist in protein_list_F)]
```

```python
unique_protein_list_S = [list(item) for item in dict.fromkeys(tuple(sublist) for sublist in protein_list_S)]
```

```python
# Read the .fasta file and create a dictionary
with open('/home/wout/databases/GUT_DB1IGC.fasta', 'r') as fasta_file:
    sequences = SeqIO.to_dict(SeqIO.parse(fasta_file, 'fasta'))

# Open the new .fasta file in write mode
with open('/home/wout/databases/F07_target_GUT_sage.fasta', 'w') as output_file:
    for sublist in unique_protein_list_F:
        protein_id = sublist[0]  # Extract the identifier from the sublist
        if protein_id in sequences:
            # Write the identifier and sequence to the new .fasta file
            SeqIO.write(sequences[protein_id], output_file, 'fasta')
```
